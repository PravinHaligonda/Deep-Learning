### 3.3 Probability Distribution

#### 3.3.1 Discreate Variable and Probability Mass Function

```python
# Probability Distribution over discreate variable is called Probability Mass Function
# Mainly Denoted by P(x = x) | P(x) and P(y = y) | P(y) and P(x=x, y=y) == P(x, y)
# 1) P(x = x) = 1 means x is certain.  2) P(x = x) = 0 means is impossible.
```

```python
dis_var = torch.tensor([0, 1, 1, 2, 3, 4, 5, 6, 7, 8])
dis_var, dis_var.nelement()
```

```python
# target value to check
target = 1

occurance = torch.sum(dis_var == target).item()

total_elements = dis_var.nelement()

prob = occurance / total_elements

# Check for probability validity
if prob < 0.0:
    raise ValueError("Probability of less than 0 is impossible for an event.")
elif prob > 1.0:
    raise ValueError("Probability greater than 1 is impossible for a single event.")
else:
    print(f"The probability of {target} in dis_var: {dis_var.tolist()}: {prob}")
print(prob)
```

```python
# 1) single discrete random variable x with k different states. We can place a uniform distribution over x.
# 2) the Probability Mass Function is 1 / k = P(x = x_i) for all i. (k is positive integer)
# 3) sum(P(x = x_i)) == sum(1 / k) == k / k == 1 the distribution is properly normalised.
```

```python
event = 1
x = torch.arange(1, 11)
k = x.nelement()
prob_mass_func = event / k
prob_mass_func
```

### 3.4 Marginal Probability

#### So This is Marginal Probability v0.01

```python
# How to find the probabilty distribution of a subset of states given a set all possible states.
```

This is for discrete variable.

```python
sample = 10
dis_random_var_x = torch.randint(sample, (1,),)
dis_random_var_y = torch.randint(sample, (1,),)
print(f"Discrete Random Variable X: {dis_random_var_x}")
print(f"Discrete Random Variable Y: {dis_random_var_y}")
```

```python
Range = torch.arange(1, sample+1)
```

```python
out = []
j = dis_random_var_x
for i in Range:
    if j in i:
        out.append(j)
    else:
        pass

prob1 = len(out) / Range.nelement()
prob1
```

```python
out = []
j = dis_random_var_y
for i in Range:
    if j in i:
        out.append(j)
    else:
        pass

prob2 = len(out) / Range.nelement()
prob2
```

```python
mar_prob = prob1 + prob2
mar_prob
```

### 3.5 Conditional Probability

```python
start, stop = 1, 11
dis_random_var_x = torch.randint(start, stop, (1,),) 
dis_random_var_y = torch.randint(start, stop, (1,),)
print(f"Discrete Random Variable X: {dis_random_var_x}")
print(f"Discrete Random Variable Y: {dis_random_var_y}")
```

```python
ss = torch.arange(start, stop)
ss
```

```python
prob = 0
if dis_random_var_x in ss:
    prob += 1
else:
    None

prob_x = prob / ss.nelement() 
```

```python
prob_x
```

```python
prob = 0
if dis_random_var_y in ss:
    prob += 1
else:
    None

prob_y = prob / ss.nelement()
```

```python
prob_y
```

```python
prob_x_and_y = prob_x + prob_y
```

```python
prob_y_given_x = prob_x_and_y / prob_x
```

### 3.6 The Chain Rule of Conditional Probability

```python
# Chain rule or Product rule of Conditional probability
# P(a, b, c) = P(a | b, c)P(b, c)
# P(b, c) = P(b | c)P(c)
# P(a, b, c) = P(a | b, c)P(b | c)P(c)
```

### 3.7 Independence and Conditional Independence

Two random variable x and y are independant if their probability distribution can be expressed as a product of two factors,
one involving only x and one involving only y

```python
# P(X = x, Y = y) = P(X = x)P(Y = y)
# event x doesn't influence event y, they are independent.
```

Two random variable x and y are conditionally independent given a random variable z if the conditional probability distribution over x and y factorizes in this way for every value of z

```python
# for all X = x, Y = y, Z = z, P(X = x, Y = y | Z = z) = P(X = x | Z = z)P(Y = y | Z = z)
# means that x and y are conditional independent given z
```

### 3.8 Expectation, Variance and Covariance

```python
dis_var = torch.tensor([1, 2, 3, 4, 4, 4, 3, 2, 1])
```

```python
unique_values, counts = torch.unique(dis_var, return_counts=True)
unique_values, counts
```

```python
total_elements = dis_var.nelement()
prob = counts.float() / total_elements
```

```python
expectation = torch.sum(unique_values.float() * prob)
expectation
```

```python
print(f"Expected value (E[x]) of dis_var: {expectation.item()}")
```
