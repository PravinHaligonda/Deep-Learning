### 3.3 Probability Distribution

#### 3.3.1 Discreate Variable and Probability Mass Function

```python
dis_var = torch.tensor([0, 1, 1, 2, 3, 4, 5, 6, 7, 8])
dis_var, dis_var.nelement()
```

```python
# Probability Distribution over discreate variable is called Probability Mass Function
# Mainly Denoted by P(x = x) | P(x) and P(y = y) | P(y) and P(x=x, y=y) == P(x, y)
# 1) P(x = x) = 1 means x is certain.  2) P(x = x) = 0 means is impossible.
```

```python
total = []
i = 1
for j in dis_var:
    if i in j:
        total.append(i)
    else:
        pass
prob = len(total) / dis_var.nelement()
if prob < 0.0:
    raise ValueError("Probability of 0 has impossible event")
elif prob > 1.0:
    raise ValueError("Probability of greater than 1 has out of state of event.")
else:
    prob
print(prob)
```

```python
# 1) single discrete random variable x with k different states. We can place a uniform distribution over x.
# 2) the Probability Mass Function is 1 / k = P(x = x_i) for all i. (k is positive integer)
# 3) sum(P(x = x_i)) == sum(1 / k) == k / k == 1 the distribution is properly normalised.
```

```python
one = 1
x = torch.arange(10)
k = x.nelement()
prob_mass_func = one / k
prob_mass_func
```

```python
probs = []
for i in x:
    j = i.nelement()
    probs.append(j / k)
    
probs = torch.tensor(probs)
print(probs[0]) # ----------------------- 2) PMF
torch.sum(probs) # ---------------------- 3) normalised to one
```
